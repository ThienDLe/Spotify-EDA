```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tracks = pd.read_csv("/Users/lethien/Desktop/Spotify Data Analysis/tracks.csv")
df_tracks.head()
```

```python
# null values
pd.isnull(df_tracks).sum()
```

```python
df_tracks.info()
```

Who has the least and the most popular songs in Spotify?

```python
sorted_df = df_tracks.sort_values('popularity', ascending=True).head()
sorted_df
```

```python
df_tracks.describe().transpose()
```

```python
most_popular = df_tracks.query('popularity>90', inplace=False).sort_values('popularity', ascending=False)
most_popular[:10]
```

```python
df_tracks.set_index('track_id', inplace=True)
df_tracks.head()
```

```python
df_tracks['duration'] = df_tracks['duration_ms'].apply(lambda x: round(x/1000))
df_tracks.drop('duration_ms', inplace=True, axis=1)
```

```python
df_tracks.duration.head()
```

Who has the most monthly listeners in Spotify?

```python
corr_df = df_tracks.drop(['key', 'mode'], axis=1).corr(method='pearson')
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(corr_df, annot=True, fmt='.1g', vmin=-1, vmax=1, center=0, cmap='inferno', linewidths=1, linecolor='Black')
heatmap.set_title('Correlation Heatmap Between Variables')
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
# since the dataset is too big, we get a small sample from the dataset (about 1%)
sample_df = df_tracks.sample(int(0.01*len(df_tracks)))
print(len(sample_df))
```

```python
plt.figure(figsize=(12,6))
sns.regplot(data=sample_df, y='loudness', x='energy', color='c').set(title='Loudness vs Energy Correlation')
```

```python
plt.figure(figsize=(12,6))
sns.regplot(data=sample_df, y='popularity', x='acousticness', color='b').set(title='Popularity vs Acousticness Correlation')
```

```python
plt.figure(figsize=(18,6))
sns.barplot(data=df_tracks, x='popularity', y='duration').set(title='Popularity vs Duration')
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(12,6))
plt.title('Duration vs Genres')
sns.color_palette('rocket', as_cmap=True)
sns.barplot(y='genre', x='duration', data=df_tracks)
plt.xlabel('Duration in seconds')
plt.ylabel('Genres')
```

```python
# top 10 most popular genres
sns.set_style(style='darkgrid')
plt.figure(figsize=(12,6))
famous = df_tracks.sort_values('popularity', ascending=False).head(10)
sns.barplot(y='genre', x='popularity', data=famous).set(title='10 most popular genres')
```
